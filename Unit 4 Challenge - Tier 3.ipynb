```python
# First, make a variable called url_LondonHousePrices, and assign it the following link, enclosed in quotation-marks as a string:
# https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls

url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"

# The dataset we're interested in contains the Average prices of the houses, and is actually on a particular sheet of the Excel file. 
# As a result, we need to specify the sheet name in the read_excel() method.
# Put this data into a variable called properties.  
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

**2.2. Cleaning the data**

You might find you need to transpose your dataframe, check out what its row indexes are, and reset the index. You  also might find you need to assign the values of the first row to your column headings  . (Hint: recall the .columns feature of DataFrames, as well as the iloc[] method).

```python
# Transpose the dataframe so that each row represents a single borough 
properties_t = properties.transpose()
properties_t.head()
```

```python
# Reset index and assign column names to the first row of the data (i.e. the dates)
properties_t = properties_t.reset_index()
properties_t.columns = properties_t.iloc[0,:]
properties_t.head()
```

```python
# Remove the first row now that the dates are saved as the column names
properties_t = properties_t.iloc[1:,:]
```

**2.3. Cleaning the data (part 2)**

You might we have to **rename** a couple columns. How do you do this? The clue's pretty bold...

```python
# Rename the columns that are not dates
properties_t = properties_t.rename(columns = {'Unnamed: 0':'London_Borough', pd.NaT:'ID'})
properties_t.head()
```

**2.4.Transforming the data**

Remember what Wes McKinney said about tidy data? 

You might need to **melt** your DataFrame here. 

```python
# Melt table so that the only columns are for London_Borough, ID, month, and avergae_price
properties_tidy = pd.melt(properties_t, id_vars=('London_Borough', 'ID'), var_name='month', value_name='Average_price')
properties_tidy.dtypes
```

Remember to make sure your column data types are all correct. Average prices, for example, should be floating point numbers... 

```python
# Convert Average_price to floats
properties_tidy['Average_price'] = pd.to_numeric(properties_tidy['Average_price'])
properties_tidy.info()
```

**2.5. Cleaning the data (part 3)**

Do we have an equal number of observations in the ID, Average Price, Month, and London Borough columns? Remember that there are only 32 London Boroughs. How many entries do you have in that column? 

Check out the contents of the London Borough column, and if you find null values, get rid of them however you see fit. 

```python
# Identify the unique London_Borough values, and which we want to keep
properties_tidy['London_Borough'].unique()
```

```python
# Investigate what Unnamed: 34, Unnamed: 37, and Unnamed: 47 look like
properties_tidy[properties_tidy['London_Borough']=='Unnamed: 34'].head()
```

```python
properties_tidy[properties_tidy['London_Borough']=='Unnamed: 37'].head()
```

```python
# Drop NaN values and check non-null count
properties_tidy = properties_tidy.dropna()
properties_tidy.info()
```

```python
# Make a list of the values that are not boroughs, and therefore we do not need for our analysis
non_borough = ['Inner London', 'Outer London', 'NORTH EAST', 'NORTH WEST', 
               'YORKS & THE HUMBER', 'EAST MIDLANDS', 'WEST MIDLANDS', 'EAST OF ENGLAND', 
               'LONDON', 'SOUTH EAST', 'SOUTH WEST', 'England']
```

```python
# Remove rows associated with the non_borough list
properties_tidy = properties_tidy[~properties_tidy['London_Borough'].isin(non_borough)]
properties_tidy['London_Borough'].unique()
```

```python
# Rename df as final as the final, clean version of the data.
df = properties_tidy
df.head()
```

**2.6. Visualizing the data**

To visualize the data, why not subset on a particular London Borough? Maybe do a line plot of Month against Average Price?

```python
# Look at Westminster as an example
westminster = df[df['London_Borough'] == 'Westminster']
westminster.plot('month', 'Average_price', ylabel='Average price')
```

To limit the number of data points you have, you might want to extract the year from every month value your *Month* column.

```python
# Add a column for year and calulate average by year
df['year'] = df['month'].apply(lambda x: x.year)
dfg = df.groupby(['London_Borough', 'year']).mean(numeric_only=True)
dfg.sample(10)
```

```python
# Reset index on the entire dataframe
dfg = dfg.reset_index()
dfg.head()
```

**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
# Create a function to calulate and return price ratio from 1998 to 2018

def create_price_ratio(d):
    year_1998 = float(d['Average_price'][d['year'] == 1998])
    year_2018 = float(d['Average_price'][d['year'] == 2018])
    ratio = year_2018 / year_1998
    
    return ratio
```

```python
# Loop through each unique borough in dfg and use create_price_ratio
# Create a dictionary with borough as the key and ratio as the value. 

borough_ratios = {}

for borough in dfg['London_Borough'].unique():
    borough_ratios[borough] = create_price_ratio(dfg[dfg['London_Borough']==borough])
    
```

```python
# Convert to dataframe
df_ratios = pd.DataFrame(borough_ratios, index=range(1))
df_ratios
```

```python
# Formatting df - transpose and rest index
df_ratios = df_ratios.transpose()
df_ratios = df_ratios.reset_index()
```

```python
# Formatting df - rename columns
df_ratios = df_ratios.rename(columns={'index':'London_Borough', 0:'ratio'})
```

```python
# sort ratios in descending order and subset the top 15 results
top15 = df_ratios.sort_values('ratio', ascending=False).head(15)
top15
```

```python
# Visualize top 15 boroughs with the greatest average home price increase from 1998 to 2018

top15.plot('London_Borough', 'ratio', kind = 'bar', 
           title = 'Top 15 London Boroughs with the greatest increase in \n average house prices from 1998 to 2018')
```
